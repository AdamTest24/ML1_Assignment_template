```python
from matplotlib.pyplot import subplots
from numpy.random import uniform, seed
from numpy import zeros, vstack
```

```python
samples, features = 20, 2

seed(123)

X1 = uniform((0, 0), (1, 1), (samples, features))
X2 = uniform((0, 0), (1, 1), (samples, features))

X = vstack((X1, X2))
y = zeros(2*samples)

y[samples:] = 1

line_pos = 0.4

plot_figure(X[:,0], X[:,1], y, line=line_pos);
```

```python
seed(11111)

split = 0.4

X1 = uniform((0, 0), (split, 1), (samples, features))
X2 = uniform((split, 0), (1, 1), (samples, features))

X = vstack((X1, X2))

y        = zeros(2*samples)

y[samples:] = 1

line_pos = 0.4

plot_figure(X[:,0], X[:,1], y, line=line_pos);
```

```python
def plot_figure(x1, x2, labels, line=False):

    fig, ax = subplots(figsize=(4,4))

    ax.scatter(x1, x2, c=labels, cmap='bwr');

    if line:
        ax.axvline(x=line, c='k')

    col = 0.9

    ax.set_facecolor((col, col, col))
    ax.set_xlabel('First Feature', fontsize=16)
    ax.set_ylabel('Second Feature', fontsize=16)
    ax.xaxis.label.set_color('k')
    ax.yaxis.label.set_color('k')

    return fig, ax
```

```python
from math import log
from numpy import asarray, nonzero
```

```python
def my_entropy(counts):
    
    probs = asarray(counts) / sum(counts)
        
    nonzero_inds = nonzero(probs)
    
    logs = [-prob*log(prob, 2) for prob in probs[nonzero_inds]]
    
    entropy = sum(asarray(logs))
    
    return entropy
```

```python
my_entropy([10, 90])
```
