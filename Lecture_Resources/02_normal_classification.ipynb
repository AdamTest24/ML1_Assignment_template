# Normally Distributed Data

```python
from numpy.random import normal, seed
```

```python
RANDOM_SEED_1  = 1234

samples, features = 100, 2

mean1, std1 = 0, 1

seed(RANDOM_SEED_1)

random_numbers1 = normal(loc=mean1, scale=std1, size=(samples, features))

print(random_numbers1.round(2))
```

```python
from matplotlib.pyplot import subplots

fig, ax = subplots()

ax.hist(random_numbers1, bins=20);
```

```python
RANDOM_SEED_2  = 12345

seed(RANDOM_SEED_2)

mean2, std2 = 3, 5

random_numbers2 = normal(loc=mean2, scale=std2, size=(samples, features))

print(random_numbers2.round(2))
```

```python
from numpy import vstack

X = vstack((random_numbers1, random_numbers2))

X.shape
```

```python
from numpy import zeros

y = zeros(X.shape[0])

y[X.shape[0]//2:] = 1

y
```

```python
fig, ax = subplots()

ax.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr');
```

# Train-Test Split

```python
from sklearn.model_selection import train_test_split

RANDOM_SEED_3 = 999

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=RANDOM_SEED_3, test_size=0.5)

X_train.shape, y_train.shape
```

```python
X_train
```

# Classification

[Random Forest Wikipedia](https://en.wikipedia.org/wiki/Random_forest)

[Decision Trees Wikipedia](https://en.wikipedia.org/wiki/Decision_tree_learning)

[Random Forest Scikit-Learn Documentation](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html)

```python
from sklearn.ensemble import RandomForestClassifier

RANDOM_SEED_4  = 111

# Instantiation of classifier
clf = RandomForestClassifier(random_state=RANDOM_SEED_4)

# Model Training
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)

y_pred.size
```

# Classification Score

```python
# Evaluating the score using test data:
score = clf.score(X_test, y_test)

score
```

# Prediction Probabilities

```python
y_pred_prob = clf.predict_proba(X_test)
```

```python
cols = 3

fig, ax = subplots(ncols=cols, figsize=(10,5))

ax[0].scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap='bwr');
ax[1].scatter(X_test[:, 0], X_test[:, 1], c=y_pred, cmap='bwr');
ax[2].scatter(X_test[:, 0], X_test[:, 1], c=y_pred_prob[:,1], cmap='bwr');

for index in range(cols):
    ax[index].set_facecolor('gray')

fig.tight_layout()
```

# Feature Importances

```python
feature_importances = clf.feature_importances_

feature_importances
```

```python
from numpy import arange

bins = arange(len(feature_importances))

fig, ax = subplots()

ax.bar(bins, feature_importances, color=('m', 'g'), width=0.3);
ax.set_xticks(bins)
ax.set_xticklabels(('Feature 1', 'Feature 2'));
```

# Predict State Space

```python
from itertools import product

from numpy import linspace, asarray, meshgrid
```

```python
dim1_min, dim1_max = -20, 20
dim2_min, dim2_max = -20, 20

resolution1, resolution2 = 10, 10

dim1 = linspace(dim1_min, dim1_max, resolution1)
dim2 = linspace(dim2_min, dim2_max, resolution2)

dim1.shape, dim2.shape
```

```python
coords = [coordinates for coordinates in product(dim1, dim2)]

coords = asarray(coords)

fig, ax = subplots()


ax.scatter(coords[:,0], coords[:,1], c='r', s=10);

ax.set_xlabel('Feature 1', fontsize=16)
ax.set_ylabel('Feature 2', fontsize=16);
```

```python
state_space_proba = clf.predict_proba(coords)

grid_shape = (resolution1, resolution2)

proba_grid = state_space_proba[:, 1].reshape(grid_shape)

contour_levels = linspace(0, 1, 6)

grid_1, grid_2 = meshgrid(dim1, dim2)

fig, ax = subplots(figsize=(4, 4))

cax = ax.contourf(grid_1, grid_2, proba_grid, cmap='Greys', levels=contour_levels)

fig.colorbar(cax)

ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap='bwr', s=10);

ax.set_xlabel('Feature 1', fontsize=16)
ax.set_ylabel('Feature 2', fontsize=16);
```
